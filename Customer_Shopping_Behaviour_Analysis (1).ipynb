```python
# Import pandas library
import pandas as pd

# Load the dataset into dataframe 'df'
df = pd.read_csv('customer_shopping_behavior.csv');
```

```python
# Show the first 5 rows of the dataset
df.head() 
```

```python
# Display dataset information (columns, data types, non-null counts)
df.info()
```

```python
# info of numerical column
df.describe()
```

```python
#include all numerical data and categorical data 
df.describe(include ='all')
```

```python
# Check the number of missing (null) values in each column
df.isnull().sum()
```

```python
# Fill missing Review Rating values with the median rating of their Category
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
# Check the number of missing (null) values in each column
df.isnull().sum()
```

```python
# Convert column names to lowercase
df.columns = df.columns.str.lower()

# Replace spaces with underscores
df.columns = df.columns.str.replace(' ', '_')

# Rename specific column
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
```

```python
df.columns
```

```python
#create a colmn age_group
labels = ['Young Adult' , 'Adult' , 'Middle-aged' , 'Senior']
df['age_group'] = pd.qcut(df['age'],q=4,labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
#create column purchase_frequency_days
frequency_mapping = {
    'Fornifhtly':14,
    'Weekly ':7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Bi-weekly': 14,
    'Annually': 365,
    'Every 3 Months':90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
df = df.drop('promo_code_used',axis=1)
```

```python
from sqlalchemy import create_engine
from urllib.parse import quote_plus

params = quote_plus(
    "DRIVER={ODBC Driver 18 for SQL Server};"
    "SERVER=localhost\\SQLEXPRESS;"
    "DATABASE=mydatabase;"
    "Trusted_Connection=yes;"
    "TrustServerCertificate=yes;"
)

engine = create_engine(f"mssql+pyodbc:///?odbc_connect={params}")

with engine.connect() as conn:
    print("SQLAlchemy Connected!")
```
